# Operators Precedence

### 1. ()            -->  Parentheses
### 2. **            -->  Exponent
### 3. *, /, //, %   -->  Multiplication, Division, Floor division, Modulus
### 4. +, -          -->  Addition, Subtraction

The Parntheses () has higher precedense that mean when we  execute the cell the first thing our compiler did is solve the equation which is in th parentheses
Then it will look for Exponents
Then it look for Multiplication or Division or Floor division or Modulus because they all have same precedense.
Then Addition or Subtraction they also have same Precedense


Note: If the 2 operator of same precedense arrived at the same time then copiler will solve the leftmost equation first.

You can also learn more about precedense of operator in python from this website
https://www.programiz.com/python-programming/precedence-associativity

```python
2/3 *4 // 4
```

# Nested For Loops

```python
for i in range(5):
    print(i)
    for j in range(3):
        print("  ", j)   # seperating print by addind ,
```

```python
for i in range(0,5):
    
    for j in range(i+1):
        print("* ",end="")
      
    # ending line after each row
    print("\r")
```
